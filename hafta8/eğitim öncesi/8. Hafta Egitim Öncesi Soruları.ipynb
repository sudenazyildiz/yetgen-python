# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız

def degistir(dosya):
    eskikelime = input("Eski kelimeyi giriniz: ")
    yenikelime = input("Yeni kelimeyi giriniz: ")

    try:
        with open(dosya, 'r', encoding='utf-8') as file:
            content = file.read()
        updated_content = content.replace(eskikelime, yenikelime)
        with open(dosya, 'w', encoding='utf-8') as file:
            file.write(updated_content)

        print(f"'{eskikelime}' kelimesi başarıyla '{yenikelime}' kelimesiyle değiştirildi.")
    except FileNotFoundError:
        print(f"Hata: '{dosya}' dosyası bulunamadı.")
    except Exception as e:
        print(f"Bir hata oluştu: {e}")

degistir(dosya)


````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, isim, hız, mileage):
        self.isim = isim
        self.hız = hız
        self.mileage = mileage

    def display_info(self):
        print(f"Name: {self.isim} Speed: {self.hız} Mileage: {self.mileage}")

class Bus(Vehicle):
    pass

school_bus = Bus("School Volvo", 180, 12)
school_bus.display_info()
````
