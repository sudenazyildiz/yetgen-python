# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import random

class Kumanda:
    def __init__(self):
        self.tv_durum = "Kapalı"
        self.ses = 0
        self.kanal_listesi = ["TRT"]
        self.aktif_kanal = "TRT"

    def tv_ac(self):
        self.tv_durum = "Açık"
        print("Televizyon açıldı.")

    def tv_kapat(self):
        self.tv_durum = "Kapalı"
        print("Televizyon kapandı.")

    def ses_ayarla(self, yeni_ses):
        self.ses = yeni_ses
        print(f"Ses seviyesi: {self.ses}")

    def kanal_ekle(self, kanal_ismi):
        self.kanal_listesi.append(kanal_ismi)
        print(f"{kanal_ismi} kanalı eklendi.")

    def kanal_sayisini_ogren(self):
        print(f"Toplam kanal sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanal(self):
        self.aktif_kanal = random.choice(self.kanal_listesi)
        print(f"Rastgele kanal: {self.aktif_kanal}")

    def tv_bilgileri(self):
        print(f"Televizyon durumu: {self.tv_durum}")
        print(f"Ses seviyesi: {self.ses}")
        print(f"Aktif kanal: {self.aktif_kanal}")
        print(f"Kanal listesi: {self.kanal_listesi}")

def main():
    kumanda = Kumanda()

    while True:
        print("""
        1. Tv Aç
        2. Tv Kapat
        3. Ses Ayarları
        4. Kanal Ekle
        5. Kanal Sayısını Öğrenme
        6. Rastgele Kanala Geçme
        7. Televizyon Bilgileri
        Çıkmak için 'q' ya basın.
        """)

        islem = input("İşlem seçiniz: ")

        if islem == 'q':
            print("Programdan çıkılıyor...")
            break
        elif islem == '1':
            kumanda.tv_ac()
        elif islem == '2':
            kumanda.tv_kapat()
        elif islem == '3':
            yeni_ses = int(input("Yeni ses seviyesi (0-100): "))
            kumanda.ses_ayarla(yeni_ses)
        elif islem == '4':
            kanal_ismi = input("Eklenecek kanal ismi: ")
            kumanda.kanal_ekle(kanal_ismi)
        elif islem == '5':
            kumanda.kanal_sayisini_ogren()
        elif islem == '6':
            kumanda.rastgele_kanal()
        elif islem == '7':
            kumanda.tv_bilgileri()
        else:
            print("Geçersiz işlem.")

if __name__ == "__main__":
    main()
```
